**Facts about Seaborn**

1.   Seaborn is a library for making statistical graphics in Python.

2.   It builds on top of matplotlib and integrates closely with pandas data structures.

3.   It is open source and free to use.

4.   Seaborn provides a number of additional visualizations, such as violin plots, box plots, and pair plots, which are not available in Matplotlib.

5.  7/10 Data professionals use Matplotlib.

```python
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv('sales_data.csv', encoding= 'ISO-8859-1')

df.dropna(inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='SALES', hue='PRODUCTLINE', palette='Set3')
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
df.dropna(inplace=True) # Pandas

plt.figure(figsize=(8, 6))
sns.histplot(data=df, y='SALES',hue='PRODUCTLINE', palette='Set3')
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(data=df, x='MONTH_ID')
plt.title('Distribution of Orders by Month')
plt.xlabel('Month ID')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a line plot for sales over time

# Sort the DataFrame by 'ORDERDATE' in ascending order
df.sort_values(by='ORDERDATE', inplace=True)

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='ORDERDATE', y='SALES')
plt.title('Sales Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='QUANTITYORDERED', y='SALES', alpha=0.5)
plt.title('Price vs. Sales')
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.show()
```

```python
df['CITY'].unique()
```

```python
# Example: Bar plot of CITY

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='CITY', y='SALES', errorbar=None)
plt.title('Sales by Territory')
plt.xlabel('Territory')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example: Box plot of CITY

plt.figure(figsize=(10, 10))
sns.boxplot(data=df, x='CITY', y='SALES', hue='YEAR_ID', palette='Set3')
plt.title('Sales by Year and City')
plt.xlabel('City')
plt.ylabel('Sales')
plt.xticks(rotation=45)  # Rotating city names for readability
plt.legend(title='City Size')
plt.show()
```

viloin-plot.png

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data=df, x='CITY', y='SALES', palette='Set3')
plt.title('Distribution of Sales by Deal Size (Violin Plot)')
plt.xlabel('Deal Size')
plt.ylabel('Sales')
plt.show()
```

```python
columns_to_include = ['SALES', 'QUANTITYORDERED', 'PRICEEACH', 'MONTH_ID']

# Create a pair plot
sns.set(style="ticks")
sns.pairplot(df[columns_to_include], diag_kind='kde', hue='MONTH_ID', palette='viridis')

plt.suptitle('Pair Plot of Sales Data', y=1.02)
plt.show()
```
