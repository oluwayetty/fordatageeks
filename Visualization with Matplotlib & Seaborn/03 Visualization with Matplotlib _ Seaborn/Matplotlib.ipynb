# **Topic 3: Bar Plots and Histograms**

```python
# Creating a bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 15, 50]
```

```python
plt.bar(categories, values, color='green', width=0.5)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
plt.barh(categories, values, color='green')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Horizontal Bar')
plt.show()
```

```python
# Error Bar Chart

x = np.arange(0, 4, 0.2)
y = np.exp(-x)

# Generate random error values (uncertainty) for each data point
e = 0.1 * np.abs(np.random.randn(len(y)))
```

```python
plt.errorbar(x, y, yerr = e, fmt = '.-')
plt.show()
```

```python
# Stacked Bar Chart

A = [15., 30., 45., 22.]
B = [15., 25., 50., 20.]

z = range(len(A))

plt.bar(z, A, color = 'b')
plt.bar(z, B, color = 'r', bottom = A)
plt.show()
```

```python
z
```

```python
data = np.random.normal(0, 5, 100)

# Create a histogram
plt.hist(data, bins=50, color='blue', alpha=0.7)

# Customize the plot
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the histogram
plt.show()
```

# **Topic 4: Pie Charts and Box Plots**


```python
plt.figure(figsize=(7,7))
x= [35, 25, 20, 20]
labels = ['Computer', 'Electronics', 'Mechanical', 'Chemical']

plt.pie(x, labels=labels)
plt.show()
```

```python
# Creating a pie chart
labels = ['Apples', 'Bananas', 'Cherries', 'Grapes']
sizes = [30, 25, 15, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['red', 'yellow', 'purple', 'green'])
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Pie Chart Example')
plt.show()
```

# **Topic 5: Advanced Plots**


```python
days = [1, 2, 3, 4, 5]
temperature = [22, 25, 28, 24, 20]

plt.plot(days,temperature)
plt.show()
```

```python
# Creating an area plot

plt.fill_between(days, temperature, color='orange', alpha=0.5, label='Temperature')
plt.xlabel('Days')
plt.ylabel('Temperature (Celsius)')
plt.title('Area Plot Example')
plt.legend()
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1, title='Subplot1')
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
```

```python
x = [10, 15, 20, 25, 30, 35, 40]
y = [20, 24, 28, 32, 36, 40, 44]

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].plot(x,y)
ax[0,0].bar(x,y) # can create multiple plots in same subplot
ax[0,1].scatter(x,y)
ax[1,0].bar(x,y)
ax[1,1].barh(x,y)

plt.show()
```

```python
# Creating a subplot with annotations
x = np.linspace(-5, 5, 100)
y = x ** 2

fig, ax = plt.subplots()
ax.plot(x, y, label='y = x^2')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Subplot Example')
ax.annotate('Minimum', xy=(0, 0), xytext=(-3, 20), arrowprops=dict(facecolor='black'))
ax.legend()
plt.show()
```

# **Topic 6: Saving and Exporting Plots**

```python
# Saving a plot to a file
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Saved Plot Example')
plt.savefig('saved_plot.png')  # The plot will be saved in the current working directory.
plt.show()
```
