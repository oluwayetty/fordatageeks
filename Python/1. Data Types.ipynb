# **Type Casting**

This is the process of converting the value of one data type(integer, string, float, etc) to another data type.

Python has two types of type conversion.
1.   Implicit Type Conversion
2.   Explicit Type Conversion



**Implicit Type Conversion**: Python converts one data type to another data types. This doesn't need any user intervention.

```python
num_int = 100
num_float = 2.9

new_num = num_int + num_float

print("Datatype of num_int is:", type(num_int))
print("Datatype of num_float is:", type(num_float))

print("Value of num_new is:", new_num)
print("Datatype of num_new is:", type(new_num))
```

```python
num_int = 100
num_str = "29"

print("Datatype of num_int is:", type(num_int))
print("Datatype of num_str is:", type(num_str))

new_num = num_int + num_str
```

**Explicit Type Conversion**: Python converts one data type to another data types. This requires user intervention.

```python
type("-34")
```

```python
int("-34")
```

```python
type(int("-34"))
```

```python
num_int = 100
num_str = "29"
num_str_to_int = int(num_str)

print("Datatype of num_int is:", type(num_int))
print("Datatype of num_float is:", type(num_str_to_int))

new_num = num_int + num_str_to_int

print("Value of num_new is:", new_num)
print("Datatype of num_new is:", type(new_num))
```
