# **Control Flow**

Control flow is the order your code runs. Usually, code runs from the first line to the last line. However, as programmers, we can control the order in which the code should run.

Python has 3 patterns to Control flow:
*  Sequential (the default pattern)
*  Selection/Decision Control for decisions and branching
*  Repetition - used for looping i.e repeating a piece of code multiple times until a condition is met.

## **Sequential**

This is the Python default way of executing a piece of code. The code is executed one after the other until the last line.

````python
num1 = 100
num2 = 34

new_num = num1 - num2
print("Value of num_new is:", new_num)
````

## **While Loop**

The code block is repeated while a particular condition is TRUE.
When the condition becomes false, the code is not longer repeated.

We use Wiile loop when we don't know what the termination condition will be or we don't know the number of iterations beforehand, we call this Indefinite Iteration.

whileloop.jpg

````python
count = 0
while count <= 3:
  print(count)
  count = count + 1
````

````python
m = 5
i = 0

while i < m:
  print(i)
  i = i + 1

print("End")
````

````python
number = int(input("Enter a number lesser than 100: "))

while number > 100:
  print("Incorrect number, enter another number")
  number = int(input("ENter a number greater than 100"))
else:
  print("Correct, {} is lesser than 100".format(number))
````

**Infinite While Loop**

````python
num = 10
while num == 10:
  print("Num is 10")
````

**Using if-else in a While Loop**

````python
num =  int(input("Enter a number"))

while num > 0 and type(num) == int:
  if num % 2 == 0:
    print("You entered an even number which is {}".format(num))
  else:
    print("You entered an odd number which is {}".format(num))
````

## **For Loop**

The code block of a For loop consists of two sections. A header specifiying the iterating conditions and a body which is executed for each iteration. The header also serve as the indicator for the terminating condition.

forloop.jpg

````python
weathers = ["Fall", "Summer", "Winter", "Spring"]
for weather in weathers:
  print(weather)
````

````python
weathers = {"Fall", "Summer", "Winter", "Spring"}
for weather in weathers:
  print(weather)
````

````python
snacks = ("Biscuits", "Chips", "Pringles")
for each_snack in snacks:
  print("I love {}".format(each_snack))
````

**For Loop with Python Range**

range() function takes in 1-3 parameters, and returns a generated list of integers from a given lower bound up to(not including) to a given upper bound. Steps in range are 1 by default, but can be increased as specified by the user.

* range(stop)
* range(start,stop)
* range(start,stop,step)

````python
for x in range(4):
  print(x)
````

````python
print(range(7))
````

````python
for x in range(3,7): #first parameter is starting number
  print(x)
````

````python
for x in range(2,10,2):
  print(x)
````

````python
for x in range(0,10,2):
  print(x)
````

````python
for x in range(0,15,5):
  print(x)
````

**Using if-else in a For Loop**

````python
number = int(input("Enter a number between 1 and 10"))

if number <= 10:
  if number % 2 == 0:
    for x in range(number):
      print(x)
  else:
    print("Number is odd number")
else:
  print("Incorrect, {} is greater than 10".format(number))
````

# **Bonus: List Comprehension**

List comprehension in Python is an easy and compact syntax for creating a list from a string or another list.

````python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)

print(newlist)
````

````python
newlist = [x for x in fruits if "a" in x]

print(newlist)
````

````python
numbers = [2,3,4,5,6]
new_nums = []

for x in numbers:
  if x % 2 == 0:
    new_nums.append(x)

print(new_nums)
````

````python
new_nums = [ x for x in numbers if x % 2 == 0]
print(new_nums)
````

````python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if x != "apple":
    newlist.append(x)

print(newlist)
````

````python
newlist = [x for x in fruits if x != "apple"]
newlist
````
