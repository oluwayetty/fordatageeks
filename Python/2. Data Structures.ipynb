## **LIST**

A list is used to store multiple items of various data types in a single variable. It is an ordered sequence of elements and it is mutable(i.e we can change, add, or remove itmes from a list after it's been created). We declare a list using the square bracket notation [].

````python
# Storing our personal information into a list
# square bracket []
personal_info = ["yetty_123", 27, 1.58, "Blueberry Muffins"]
print(personal_info)
````

````python
type([])
````

**Verify it's a List**

````python
print(type(personal_info))
````

**Can lists have duplicates?** YES


````python
# Storing our different data types in a list
personal_info = [True, 12.5, "yetty_123", 1.58, False, "Blueberry Muffins", 27, True, "yetty_123"]
print(personal_info)
````

**Lists are Ordered and can be Indexed by their Order** # show a list diagram with index

List Items are indexed, the first item has index [0], second items has index [1] and so on.

````python
print(personal_info[0])
print(personal_info[3])
print(personal_info[8])
print(personal_info[-1])

print(personal_info[-3])
# print(personal_info[9])
````

````python
print(personal_info[0:2])
print(personal_info[3:6])
````

**Operations (i.e Functions) we can perform on a List**

````python
print(dir(personal_info))
````

\**Functions**

* append() - Adds an element at the end of the list
* clear() - Removes all the elements from the list
* copy() - Returns a copy of the list
* count() - Returns the number of elements with the specified value
* extend() - Add the elements of a list (or any iterable), to the end of the current list
* index() - Returns the index of the first element with the specified value
* insert() - Adds an element at the specified position
* pop() - Removes the element at the specified position
* remove() - Removes the item with the specified value
* reverse() - Reverses the order of the list
* sort() - Sorts the list italicised text

````python
countries = [ "United Kingdom", "Japan", "Nigeria"]
print(countries)
````

````python
countries.append("Italy")
print(countries)
````

````python
countries.remove("Japan")
print(countries)
````

````python
countries.insert(1, "UAE")
print(countries)
````

````python
numbers = countries.count("UAE")
print(numbers)
````

````python
countries.insert(0, "Nigeria")
print(countries)
````

````python
numbers = countries.count("Nigeria")
print(numbers)
````

````python
countries.clear()
print(countries)
````

## **DICTIONARY**

A typical dictionary have words and the meaning of the words ordered alphabetically. No two words are repeated in a Dictionary, every word is unique. E.g

https://img-cache.oppcdn.com/fixed/4018/assets/LHOuqdOxmOelBngz.jpg

It is the same with Python dictionaries. A word and its meaning. However, we call it a Key(word) and it's value(meaning). This is why it's generally known as an associative array because of its key-values pairs where eack key value maps to its associated value.

```
d = {
  <key>: <value>,
  <key>: <value>,
  .
  .
  .
  <key>: <value>
}
```
Dictionaries are changeable, **but do not allow duplicates**. As of Python 3.7, dictionaries are ordered. In Python 3.6 and earlier,dictionaries are unordered.

````python
dict_info = {
    "name": "yetty",
    "year_of_birth": 1995,
    "height": 1.74,
    "loves_pet?": True
}
print(dict_info)
````

**Let's test NON-DUPLICATES**

````python
dict_info = {
    "year_of_birth": 1995,
    "name": "yetty",

    "height": 1.74,
    "loves_pet": True,
    "height": 1.83
}

print(dict_info)
````

**Verify it's a Dictionary**

````python
print(type(dict_info))
````

**Operations (i.e Methods) we can perform on a Dictionary**

````python
print(dir(dict_info))
````

**Functions**

* clear() - Removes all the elements from the dictionary
* copy() - Returns a copy of the dictionary
* fromkeys() - Returns a dictionary with the specified keys and value
* get() - Returns the value of the specified key
* items() - Returns a list containing a tuple for each key value pair
* keys() - Returns a list containing the dictionary's keys
* pop() - Removes the element with the specified key
* popitem() - Removes the last inserted key-value pair
* setdefault() - Returns the value of the specified key. If the key does not * exist: insert the key, with the specified value
* update() - Updates the dictionary with the specified key-value pairs
* values() - Returns a list of all the values in the dictionary

**Indexing in Dictionaries**: Indexing in Dictionaries is done through the Keys


````python
print(dict_info['name'])
````

````python
print(dict_info.get("name"))
````

````python
print(dict_info.keys())
````

````python
print(dict_info.values())
````

````python
dict_info.update({"best_color": "White"})
print(dict_info)
````

````python
dict_info.popitem() # removes last item
print(dict_info)
````

````python
dict_info.pop("height")
print(dict_info)
````
